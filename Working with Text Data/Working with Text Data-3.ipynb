```python
import pandas as pd
```

```python
data = pd.read_csv('chicago.csv')
data['Department'] = data['Department'].astype('category')
data.dropna(inplace = True, how = 'all')
data.head()
```

```python
data['Name'].str.split(",").str.get(0)
```

```python
data['Name'].str.split(',').str.get(1)
```

```python
data['Name'].str.split(',').str.get(1).str.split(' ')
```

```python
data['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0).value_counts().head(5)
```

### The expand and n Parameters of the str.split() Method

```python
data['Name'].str.split(',', expand = False)
```

```python
data[['First Name', 'Last Name']] = data['Name'].str.split(',', expand = True)
data.head()
```

```python
data['Position Title'].str.split(' ', expand = True)
```

```python
data[['First Title Word', 'Remaining Words']] = data['Position Title'].str.split(' ', expand = True, n = 1)
data.head()
```
