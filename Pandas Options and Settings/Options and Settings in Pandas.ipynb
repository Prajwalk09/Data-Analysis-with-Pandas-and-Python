### Changing Options and Attributes and Dot Syntax

```python
import pandas as pd
import numpy as np
```

```python
data = np.random.randint(low = 0, high = 100, size = [1000, 50])
```

```python
df = pd.DataFrame(data = data)
df.head()
```

```python
pd.options.display.max_rows
```

```python
pd.options.display.max_rows = 18
pd.options.display.max_columns = 30
df
```

```python
pd.describe_option('max_rows')
```

```python
pd.describe_option('max_columns')
```

```python
pd.reset_option('max_rows')
```

### The precision Option

```python
df = pd.DataFrame(np.random.randn(5,5))
df
```

```python
pd.options.display.precision = 2
"""Rounds numbers to 2 decimal places"""
df
```

```python
pd.reset_option('precision')
"""Resets the precision option back to its original value"""
df
```
