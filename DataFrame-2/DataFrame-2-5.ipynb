### drop_duplicates() Method

```python
import pandas as pd
```

```python
data = pd.read_csv('employees.csv', parse_dates = ['Start Date', 'Last Login Time'])
data['Gender'] = data['Gender'].astype('category')
data['Team'] = data['Team'].astype('category')
data['Senior Management'] = data['Senior Management'].astype('category')
data.head()
```

```python
data.sort_values(by = ['First Name'], inplace = True)
```

```python
len(data)
```

```python
len(data.drop_duplicates())
# Only removes rows, if values for 2 or more rows are same across all columns
```

```python
data.drop_duplicates(subset = ['First Name'], keep = 'first')
data.drop_duplicates(subset = ['First Name'], keep = 'last')
data.drop_duplicates(subset = ['First Name'], keep = False)
```

```python
data.drop_duplicates(subset = ['Team'], keep = False)
```

```python
data[data.duplicated(subset = ['First Name', 'Team'])]
```
