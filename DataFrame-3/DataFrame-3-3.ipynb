### Deleting Rows or Columns from a DataFrame

```python
import pandas as pd
```

```python
data = pd.read_csv('jamesbond.csv')
data = data.set_index(keys = ['Film']).sort_index(ascending = True)
data.head()
```

```python
data.drop(labels = ['A View to a Kill', 'Casino Royale'], axis = 'rows')
new_data = data.drop(['Box Office', 'Bond Actor Salary', 'Actor'], axis = 'columns')
```

```python
print(type(new_data))
new_data
```

```python
actor = data.pop('Actor')
```

```python
actor.value_counts()
```
