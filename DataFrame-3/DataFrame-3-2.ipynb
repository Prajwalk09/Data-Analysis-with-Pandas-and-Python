### Second Argument to loc[] and iloc[]

```python
import pandas as pd
```

```python
data = pd.read_csv('jamesbond.csv')
data = data.set_index(keys = 'Film').sort_index()
data.head()
```

```python
data.loc['Die Another Day']
```

```python
data.loc['Die Another Day', "Actor"]
data.loc['Die Another Day', "Director"]
data.loc['Die Another Day', ["Actor", "Director", "Box Office"]]
data.loc[['Die Another Day', 'Casino Royale'], ["Actor", "Director", "Box Office"]]
```

```python
data.loc["Die Another Day", "Director":"Budget"]
data.loc["Casino Royale" : "Die Another Day", "Director":"Budget"]
```

```python
data.iloc[14, 2]
data.iloc[14, 2:5]
data.iloc[[14,17], 2:5]
```

```python
data.iloc[[14,18], [2,4]]
```

### Setting New Value for a Specific Cell

```python
data.loc['Dr. No', 'Actor'] = 'Sir Sean Connery'
data.loc['Dr. No']
```

```python
data.loc['Dr. No', ['Box Office', 'Budget', 'Bond Actor Salary']] = [448000000, 7000000, 600000]
data.loc['Dr. No']
```

### Setting Multiple Values

```python
actor_is_sean_connery = data['Actor'] == 'Sean Connery'
data.loc[actor_is_sean_connery, 'Actor'] = 'Sir Sean Connery'
```

```python
data.loc[data['Actor'] == 'Sir Sean Connery']
```

### Rename Index Labels or Columns in a DataFrame

```python
data.rename(mapper = {"GoldenEye": 'Golden Eye', 
                      'The World Is Not Enough':'Best Bond Movie'})
data.rename(mapper = {"GoldenEye": 'Golden Eye', 
                      'The World Is Not Enough':'Best Bond Movie'}, axis = 'index')
data.rename(index = {"GoldenEye": 'Golden Eye', 
                      'The World Is Not Enough':'Best Bond Movie'}, inplace = True)
```

```python
data.head(1)
```

```python
data.rename(mapper = {"Year" : "Release Date", "Box Office": "Revenue"}, axis = 'columns')
data.rename(columns = {"Year" : "Release Date", "Box Office": "Revenue"}, inplace = True)
data.head()
```

```python
data.columns = ['Year of Release', 'Actor', 'Director', 'Gross', 'Cost', 'Salary']
data.head()
```
