### sort_values() Method

```python
import pandas as pd
```

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

```python
nba.sort_values(by = ['Team'], ascending = False)
nba.sort_values(by = ['Team','Name'], ascending = [False, True], inplace = True)
```

```python
nba.sort_index()
```

```python
nba.sort_values(by = ['Salary'], ascending=False, na_position='last')
```

```python
nba.sort_index(inplace = True)
```

### rank() Method

```python
nba.dropna(how = 'all')
nba['Salary'] = nba['Salary'].fillna(value = 0).astype(int)
```

```python
nba['Salary'].rank()
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending = False).astype(int)
nba.head()
```

```python
nba.sort_values(by = ['Salary'], ascending = False)
```
