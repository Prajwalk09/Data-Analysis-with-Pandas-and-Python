### Selecting Multiple Columns from a DataFrame

```python
import pandas as pd
```

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

```python
nba[['Name', 'Team']]
nba[['Salary', 'Team', 'Name']].tail()
```

```python
required_columns = ['Salary', 'Team', 'Name']
nba[required_columns]
```

### Adding New Columns to a DataFrame

```python
nba.head()
```

```python
nba['Sport'] = "Basketball"
```

```python
nba['League'] = 'National Basketball Association'
nba.head()
```

```python
nba = pd.read_csv('nba.csv')
nba.head()
```

```python
nba.insert(loc = 3, column = 'Sport', value = 'Basketball')
nba.head()
```

```python
nba.insert(loc = len(nba.columns), column = 'League', value = 'National Basketball Association')
nba.head()
```
