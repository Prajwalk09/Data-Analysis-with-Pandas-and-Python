### The transpose() Method

```python
import pandas as pd
```

```python
data = pd.read_csv('bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country']).sort_index()
data.head()
```

```python
data = data.transpose()
```

```python
data.head()
```

```python
data.loc[("Price in US Dollars",), ('2010-01-01', 'Australia')]
data.loc[("Price in US Dollars",), ('2010-01-01', 'Australia') : ('2010-01-01', 'Ukraine')]
```

### Using the swaplevel() Method

```python
data = pd.read_csv('bigmac.csv', parse_dates = ['Date'], index_col = ['Date', 'Country']).sort_index()
data.head()
```

```python
data.swaplevel()
```

```python
data.swaplevel("Date", 'Country')
data.swaplevel("Country", 'Date')

data.swaplevel(0,1)
data.swaplevel(1,0)

'''All of them perform the same result of swapping the two levels.'''
```

### stack() Method

```python
data = pd.read_csv('worldstats.csv', index_col = ['country', 'year'])
data.head()
```

```python
'''Doubles the number of rows but halves the number of columns in the DataFrame'''

pd.DataFrame(data.stack(), columns = ['Column'])
```

```python
data.stack().loc[('Arab World', 2013, 'Population')]
```

```python
data.stack().to_frame().rename(columns = {0:'Column'})
```

### Arguments to the unstack() Method

```python
data = pd.read_csv('worldstats.csv')
data.set_index(keys = ['country', 'year'], inplace = True)
data.head()
```

```python
stacked_data = data.stack()
stacked_data.head(6)
```

```python
stacked_data.head(6).unstack()
```

```python
stacked_data.head(6).unstack().unstack()
```

```python
"""The country column has an index of 0. The year column has an index of 1 and third column 
which does not have a name has the index of 2"""
stacked_data
```

```python
"""The operations 
stacked_data.unstack() and stacked_data.unstack(level = 2) perform the same operation 
i.e, unstacking the column at index value 2 (the column without a name)"""

"""The unstack() Method can take integer values as arguments for the level parameter and this 
parameter defines the level at which the unstacking operation has to applied."""

stacked_data.unstack()
```

```python
stacked_data.unstack(level = 2)
```

```python
stacked_data
```

```python
stacked_data.unstack(level = 0)
```

```python
stacked_data.unstack(level = 1)
```

```python
"""Negative indexing works similar as in case of lists or any other slicing operations"""
stacked_data.unstack(level = -1)
stacked_data.unstack(level = -2)
```

```python
"""If indexes have names then those names can be used for unstacking"""
stacked_data.unstack(level = 'year')
```

```python
stacked_data.unstack(level = 'country').head()
```
