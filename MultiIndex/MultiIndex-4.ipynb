### The unstack() Method

```python
import pandas as pd
```

```python
data = pd.read_csv('worldstats.csv', index_col = ['country', 'year'])
data.head()
```

```python
stacked_data = data.stack()
stacked_data.head()
```

```python
"""The unstack method can also take lists of indexes or strings as argument to the level parameter"""
'''In the stacked_data country has index position 0, year has index position 1 and the third column has index 
position 2.'''

stacked_data.unstack(level = [1,0])
```

```python
stacked_data.unstack(level = [0,1])
```

```python
stacked_data.unstack(level = ['year', 'country'])
```

```python
stacked_data.unstack(level = ['country', 'year'])
```

```python
stacked_data.unstack(level = 'year' , fill_value = 'Missing Value').head()
```

```python
stacked_data.unstack(level = 'year' , fill_value = 0).head()
```

```python
stacked_data.unstack(level = ['country','year'])
```

### The pivot() Method

```python
data = pd.read_csv('salesmen.csv', parse_dates = ['Date'])
data['Salesman'] = data['Salesman'].astype('category')
data.head()
```

```python
data.info()
```

```python
data.pivot(index = ['Date'], columns = ['Salesman'], values = ['Revenue']).head()
```

```python
data[data['Date'] == '2016-01-01']
```

### The pivot_table() Method

```python
data = pd.read_csv('foods.csv')
data.head()
```

```python
data.pivot_table(values = ['Spend'] , index = ['Gender'] , aggfunc = 'mean')
```

```python
data.pivot_table(values = ['Spend'] , index = ['Gender','Item'] , aggfunc = 'sum', columns = ['City'])
```
