### Input and Output

```python
import pandas as pd
```

```python
url = "https://data.cityofnewyork.us/resource/25th-nujf.csv"
data = pd.read_csv(url)
data.head()
```

### Export Excel File

```python
url = 'https://data.cityofnewyork.us/resource/25th-nujf.csv'
baby_names = pd.read_csv(url)
baby_names.columns = ['Birth Year', "Gender", "Ethnicity", "Child's First Name", "Count", "Rank"]
baby_names.head()
```

```python
girls = baby_names[baby_names['Gender'] == 'FEMALE']
boys = baby_names[baby_names['Gender'] == 'MALE']
```

```python
"""Steps to write to an Excel File
1) Create an ExcelWriter object specifiying the name of the Excel sheet to be created.
2) On the DataFrame which is to be inserted into one of the sheets of the excel file, pass the ExcelWriter object 
   which was created earlier, along with few other arguments.
3) Call the save() method on the ExcelWriter object."""
```

```python
excel_file = pd.ExcelWriter("Baby_Names.xlsx")
```

```python
girls.to_excel(excel_writer = excel_file, sheet_name = "Girls", index = False)
```

```python
boys.to_excel(excel_writer = excel_file, sheet_name = "Boys", index = False, columns = boys.columns[:4])
```

```python
excel_file.save()
```
